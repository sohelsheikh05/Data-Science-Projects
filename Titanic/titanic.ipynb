```python
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.impute import SimpleImputer
class AgeImputer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X.copy()
        imputer = SimpleImputer(strategy='mean')
        # Flatten result to 1D
        X['Age'] = imputer.fit_transform(X[['Age']]).ravel()
        return X

```

```python
class FeatureEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X.copy()

        # One-hot encode 'Embarked'
        embarked_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
        embarked_matrix = embarked_encoder.fit_transform(X[['Embarked']])
        embarked_cols = embarked_encoder.get_feature_names_out(['Embarked'])
        embarked_df = pd.DataFrame(embarked_matrix, columns=embarked_cols, index=X.index)
        X = pd.concat([X, embarked_df], axis=1)

        # One-hot encode 'Sex'
        sex_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
        sex_matrix = sex_encoder.fit_transform(X[['Sex']])
        sex_cols = sex_encoder.get_feature_names_out(['Sex'])
        sex_df = pd.DataFrame(sex_matrix, columns=sex_cols, index=X.index)
        X = pd.concat([X, sex_df], axis=1)

        return X
```

```python
from sklearn.preprocessing import OneHotEncoder
class FeatureDropper(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        return X.drop(['Embarked', 'Name', 'Ticket', 'Cabin', 'Sex', 'N'], 
                      axis=1, errors='ignore')
```

```python
from sklearn.pipeline import Pipeline
pipeline=Pipeline([('ageImputer',AgeImputer()),('featureencoder',FeatureEncoder()),('featuredropper',FeatureDropper())])
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
models = {
    "logistic": LogisticRegression(),
    "knn": KNeighborsClassifier(),
    "naive": GaussianNB(),
    "decision": DecisionTreeClassifier(),
    "svc": SVC(probability=True),
    
}
```

```python
param_grid = {
    "logistic": {
        "C": [0.01, 0.1, 1, 10, 100],        # regularization strength
        "penalty": ["l1", "l2", "elasticnet", "none"], 
        "solver": ["lbfgs", "saga"],         # saga handles elasticnet
        "max_iter": [200, 500]
    },
    
    "knn": {
        "n_neighbors": [3, 5, 7, 9, 11],
        "weights": ["uniform", "distance"],
        "metric": ["euclidean", "manhattan", "minkowski"]
    },
    
    "naive": {
        # GaussianNB has only var_smoothing
        "var_smoothing": [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
    },
    
    "decision": {
        "criterion": ["gini", "entropy", "log_loss"],
        "max_depth": [None, 5, 10, 20, 30],
        "min_samples_split": [2, 5, 10],
        "min_samples_leaf": [1, 2, 4]
    },
    
    "svc": {
        "C": [0.1, 1, 10, 100],
        "kernel": ["linear", "poly", "rbf", "sigmoid"],
        "gamma": ["scale", "auto"]
    },
   
}

import warnings
warnings.filterwarnings("ignore")
```

```python
result=[]
for name,model in models.items():
    grid_search=GridSearchCV(model,param_grid[name],cv=5,scoring='accuracy',return_train_score=True)
    grid_search.fit(X_data,Y_data)
    result.append({
        'model':name,
        'accuracy':grid_search.best_score_,
        'parameters':grid_search.best_params_
    })
```
